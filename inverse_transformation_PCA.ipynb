```python
import sklearn
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_digits
digits = load_digits() # load the dataset
print(digits.data.shape) # dataset object has data which is ndarray of shape(1797,64) meaning each image has 64 pixels ( 8x8)
```

```python
df=pd.DataFrame(digits.data) #convert the ndarray to dataframe
df
```

```python
arr = df.iloc[:10].to_numpy(dtype=float)
print(arr.shape)
print(arr)
print(type(arr))
```

```python
num_images = len(arr)
num_cols = 5  # Number of columns in the grid
num_rows = (num_images + num_cols - 1) // num_cols  # Automatically calculate the number of rows based on the number of images
fig, axes = plt.subplots(num_rows, num_cols, figsize=(12, 8))
for i, ax in enumerate(axes.flat):
    if i < num_images:
        ax.matshow(arr[i].reshape((8, 8)), cmap='gray')
        ax.set_title(f'Image {i+1}')
        ax.axis('off')
    else:
        ax.axis('off')

plt.tight_layout()
plt.show()
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=0.95)
x_reduced = pca.fit_transform(df) # already and array
print(x_reduced.shape)
```

As we can see, we have reduced down the dimensions of dataset from 64 to 29 which is around 45% reduction.
Lets, get plot the first image of the decompressed dataset ( i.e back to 64 )

```python
x_recovered = pca.inverse_transform(x_reduced)
x_recovered.shape
arr2 = x_recovered[0:10]
arr2.shape
```

```python
num_images = len(arr2)
num_cols = 5  # Number of columns in the grid
num_rows = (num_images + num_cols - 1) // num_cols  # Automatically calculate the number of rows based on the number of images
fig, axes = plt.subplots(num_rows, num_cols, figsize=(12, 8))
for i, ax in enumerate(axes.flat):
    if i < num_images:
        ax.matshow(arr2[i].reshape((8, 8)), cmap='gray')
        ax.set_title(f'Image {i+1}')
        ax.axis('off')
    else:
        ax.axis('off')

plt.tight_layout()
plt.show()
```
